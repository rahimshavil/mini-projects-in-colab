```python
queue=[]

print("--- Pushing items---")
queue.append("A")
print("queue:", queue)

queue.append("Bo")
print("queue:", queue)

queue.append("Book C")
print("queue:", queue)

if queue:
    print(f"\nTop item is: {queue[-1]}")


print("\n--- Popping items ---")


removed_item = queue.pop(0)
print(f"Removed item: {removed_item}")
print("queue:", queue)

removed_item = queue.pop(0)
print(f"Removed item: {removed_item}")
print("queue:", queue)

removed_item  = queue.pop(0)
print(f"Removed item: {removed_item}")
print("queue:", queue)

if not queue:
    print("\nQueue is now empty.")







```
